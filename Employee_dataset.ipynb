```python
plt.figure(figsize=(12, 6))
sns.countplot(data=emp, x='Department')
plt.title('number of employees in each department')
plt.xticks(rotation=90)
plt.show()
```

**Employee Distribution by Department**

*There is a significant variation in the number of employees across different departments, indicating varying levels of resource allocation. Quality Control, Major Mfg Projects, and Sales have the highest number of employees, suggesting core operational functions or high demand for their services. Departments like Green Building, Facilities/Engineering, and Training have a relatively smaller employee count, which might indicate specialized roles or support functions.*

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Country', y='Monthly Salary', data=emp)
plt.title('Monthly Salary Distribution Across Countries')
plt.show()
```

**Monthly Salary Distribution Across Countries**

*There is a notable difference in median salaries across the countries. Saudi Arabia and the United Arab Emirates appear to have higher median salaries compared to Egypt, Syria, and Lebanon. The spread of salaries (as indicated by the box height) varies across countries. Some countries, like Saudi Arabia and the United Arab Emirates, exhibit a wider range of salaries, suggesting a higher degree of income inequality.The presence of outliers, as represented by the whiskers and individual data points, suggests the existence of extremely high salaries in some countries. The box plots indicate that salary distributions are generally right-skewed, with a larger concentration of lower salaries and a smaller number of high salaries.*

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Department', y='Sick Leaves', data=emp)
plt.xticks(rotation=90)
plt.title('Sick Leaves by Department')
plt.show()
```

**Sick Leaves by Department**

*There is significant variation in the number of sick leaves across departments, with some departments reporting substantially higher levels than others. Some departments have a notably higher number of sick leaves compared to the overall average, indicated by the length of the error bars. The underlying reasons for these variations could be attributed to factors such as workload, job stress, working conditions, or industry-specific hazards.*
